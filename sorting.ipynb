```python
#naive - 3vars count nos - 2traversals  => TC - ? and SC -? 

#sort 0,1,2 - dutch national flag algo 
#https://leetcode.com/problems/sort-colors/ 

def swap(a,b):
    t=a
    a=b
    b=t
    return a,b

def sort012(arr):
    l,m,h=0,0,len(arr)-1
    #3 contd 
    while(m<=len(arr)-1):
        if arr[m]==0:
            arr[m],arr[l]=swap(arr[m],arr[l])
            l+=1
            m+=1
        elif arr[m]==1:
            m+=1
        else:#if 2 then place at h and dec h 
            arr[m],arr[h]=swap(arr[m],arr[h])
            h-=1#dont move mid as it may b 0 so u need to swap again it with 1 
            #since m wont be 
            print(h)
    return arr      

sort012([0,1,1,0,1,2,1,2,0,0,0,1])   
```


The reason why you should not use while(m<=len(arr)-1) as the loop condition in the sort012 function is that it can result in an "IndexError: list index out of range" error.

Here's why: when m equals len(arr)-1, the code inside the while loop will be executed, and the if statement that checks if arr[m] is equal to 0, 1, or 2 will be executed. If arr[m] is equal to 0 or 1, the code will enter one of the first two if statements and modify the array by swapping elements or incrementing m. If arr[m] is equal to 2, the code will enter the else statement and swap arr[m] with arr[h], and decrement h.

The problem is that after the swap in the else statement, the value of m does not change. Therefore, if arr[m] is swapped with arr[h], the value of arr[m] will be 2, but the next iteration of the loop will still check the same index m, which may now be out of bounds if h was already at index m. This can cause an "IndexError: list index out of range" error to be raised, since arr[m] is accessed but m is now larger than the highest index in the array.

In contrast, using the condition while m <= h as the loop condition in the sort012 function ensures that m will never be greater than h, so there will be no risk of an "IndexError: list index out of range" error.
