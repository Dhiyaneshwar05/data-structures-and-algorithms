```python
#naive - 3vars count nos - 2traversals  => TC - ? and SC -? 

#sort 0,1,2 - dutch national flag algo 
#https://leetcode.com/problems/sort-colors/ 

def swap(a,b):
    t=a
    a=b
    b=t
    return a,b

def sort012(arr):
    l,m,h=0,0,len(arr)-1
    #3 contd 
    while(m<=h):                     #len(arr)-1): ===> analyse y this contd wont give desired results 
        if arr[m]==0:
            arr[m],arr[l]=swap(arr[m],arr[l])
            l+=1
            m+=1
        elif arr[m]==1:
            m+=1
        else:#if 2 then place at h and dec h 
            arr[m],arr[h]=swap(arr[m],arr[h])
            h-=1#dont move mid as it may b 0 so u need to swap again it with 1 
            #since m wont be 
            print(h)
    return arr      

sort012([0,1,1,0,1,2,1,2,0,0,0,1])   
```


The reason why you should not use while(m<=len(arr)-1) as the loop condition in the sort012 function is that it can result in an "IndexError: list index out of range" error.

Here's why: when m equals len(arr)-1, the code inside the while loop will be executed, and the if statement that checks if arr[m] is equal to 0, 1, or 2 will be executed. If arr[m] is equal to 0 or 1, the code will enter one of the first two if statements and modify the array by swapping elements or incrementing m. If arr[m] is equal to 2, the code will enter the else statement and swap arr[m] with arr[h], and decrement h.

The problem is that after the swap in the else statement, the value of m does not change. Therefore, if arr[m] is swapped with arr[h], the value of arr[m] will be 2, but the next iteration of the loop will still check the same index m, which may now be out of bounds if h was already at index m. This can cause an "IndexError: list index out of range" error to be raised, since arr[m] is accessed but m is now larger than the highest index in the array.

In contrast, using the condition while m <= h as the loop condition in the sort012 function ensures that m will never be greater than h, so there will be no risk of an "IndexError: list index out of range" error.

What are the Boundary Cases of the Insertion Sort algorithm?
Insertion sort takes maximum time to sort if elements are sorted in reverse order. And it takes minimum time (Order of n) when elements are already sorted. 

What are the Algorithmic Paradigm of Insertion Sort algorithm?
Insertion Sort algorithm follows incremental approach.

Is Insertion Sort an in-place sorting algorithm?
Yes, insertion sort is an in-place sorting algorithm.

Is Insertion Sort a stable algorithm?
Yes, insertion sort is a stable sorting algorithm.

When is the Insertion Sort algorithm used?
Insertion sort is used when number of elements is small. It can also be useful when input array is almost sorted, only few elements are misplaced in complete big array.

What is Binary Insertion Sort? 
We can use binary search to reduce the number of comparisons in normal insertion sort. Binary Insertion Sort uses binary search to find the proper location to insert the selected item at each iteration. In normal insertion, sorting takes O(i) (at ith iteration) in worst case. We can reduce it to O(logi) by using binary search. The algorithm, as a whole, still has a running worst case running time of O(n^2) because of the series of swaps required for each insertion. 1. sel min and swap it with 1st index 
2. next 2nd min at 2nd index / min in the remaining part of array (ie) after 1st index 
3. // example - say 6 elem in arr => then 5 steps last ele no need to sort


TC -> O(N^2)

```python
#selection sort 

#by me - TC ? 
#how to opti ?
#used many - built ins -> index(), min()

def sel_sort_naive(arr):
    #arr=[13,46,24,52,20,9]


    for i in range(len(arr)-1): #n-1 steps
        min_e_i=arr.index(min(arr[i:])) # got min ele 
        arr[i],arr[min_e_i]=arr[min_e_i],arr[i] #swapping

    return(arr)

sel_sort_naive([13,46,-1,0,24,52,20,9])
```

```python
def sel_sort(arr):
    
    for i in range(0,len(arr)-1): #uptill n-1 ele as last ele sort itself     
        min_i=i
        
        #to find min ele in range 
        for j in range(i,len(arr)):
            if arr[min_i]>arr[j]:
                min_i=j
        
        #after finding min - swap 
        arr[min_i],arr[i]=arr[i],arr[min_i]
        
    return arr

sel_sort([13,46,-1,0,24,52,20,9])
```

```python
# times - select from 1st then insert in correct position after elememnts will be swapped till it can 
#[13,46,-1,0,24,52,20,9]

def ins_sort_naive(arr):
    for i in range(1,len(arr)):# to take element 
        
        j=i-1 #to search till the bfore entry 
        
        while(j>=0 and arr[j]>arr[j+1]):#if any element is higher and to the left of it 
            #thn 1 shift to the right position 
            
            arr[j],arr[j+1]=arr[j+1],arr[j]
            j-=1
            
    return arr
ins_sort_naive([-10,46,-1,0,24,52,20,9])#[0,34,78,119])#[-10,46,-1,0,24,52,20,9])
```
