```python
#https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/description/

# min in array - O(logN)
def min_bs(nums):
    l=0
    h=len(nums)-1
    m=(l+h)//2
    while(l<h):
        if nums[m]<nums[h]:
            h=m
            m=(l+h)//2
        else:
            l=m+1
            m=(l+h)//2
    return l

#1
a=min_bs([14,15,20,22,25,1,3,5,7,9,11])
print("min at index-",a,"\tval is-",[14,15,20,22,25,1,3,5,7,9,11][a])

#2
a=min_bs([9,11,14,15,20,22,25,1,3,5,7])
print("min at index-",a,"\tval is-",[9,11,14,15,20,22,25,1,3,5,7][a])

#3
a=min_bs([1,3,5,7,9,11,14,15,20,22,25])
print("min at index-",a,"\tval is-",[1,3,5,7,9,11,14,15,20,22,25][a])
```
