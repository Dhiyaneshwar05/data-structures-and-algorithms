```python
#stack using list()
class Stack:
    def __init__(self):
        self.buffer=list()
    
    def push(self,val):
        self.buffer.append(val)
    
    def pop(self):
        self.buffer.pop()
    
    def isempty(self):
        return len(self.buffer)==0
    
    def size(self):
        return len(self.buffer)
    
    
s1=Stack()
```

```python
s1.push(12)
s1.push(2)
s1.push(24)
```

```python
s1.pop()
s1.pop()
s1.pop()
```

```python
s1.isempty()
```

```python
class stack_arr: #using arr, top var => stack 
    
    def __init__(self):
        self.buffer=list()
        self.top=-1
    def push(self,val):
        self.top+=1
        self.buffer.insert(self.top,val)
    def pop(self):
        self.buffer.pop()
        self.top-=1
    def size(self):
        return len(self.buffer)
    def isempty(self):
        return len(self.buffer)==0

s2=stack_arr()
```

```python
s2.push(12)
s2.push(2)
s2.push(24)
```

```python
s2.size()
```

```python
s2.pop()
s2.pop()
s2.pop()
```

```python
s2.isempty()
```

```python
# VALID PARENTHESIS 
# https://leetcode.com/problems/valid-parentheses/ 


def isValid(s):
    
    if len(s)%2==1:
        return False 
    
    else:
        #pairs 
        d={'}':'{',')':'(',']':'['}
        
        stk=[]
        
        for i in s:
            #open - add in stk 
            if i in d.values():
                stk.append(i)
            else:
                #means stk is empty when closing comes 
                if len(stk)==0:
                    return False 
                #take top ele in stack and find if it matches pair
                elif d[i]==stk.pop():
                    pass 
                else:
                    return False 
        #if empty then True, else False --- TstC- "(("
        return not(len(stk))
    
        
isValid("[[{()}]{}()]")
```
