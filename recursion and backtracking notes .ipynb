```python
# functional - broken down into smol parts 

def sum2(n):
    if n==0:
        return 0 
    return n+sum2(n-1)
sum2(5)
```

```python
#factorial N
def fact(n):
    if n==1:
        return 1
    return n*fact(n-1)
fact(7)
```

```python
#rev a list 
a=[1,12,23,4]
def rev(l,r):
    if l>r:
        return 
    t=a[l]
    a[l]=a[r]
    a[r]=t
    rev(l+1,r-1)
```

```python
rev(0,3)
a
```

```python
def one_var_rec(i):
    if i>=len(b)/2:
        return 
    t=b[i]
    b[i]=b[len(b)-i-1]
    b[len(b)-i-1]=t
    
    one_var_rec(i+1)
one_var_rec(0)
```

```python
s="MADaAM"
def palindrome(i):
    if i>len(s)/2:
        return True
    if(s[i]!=s[len(s)-1-i]):
        return False 
    return palindrome(i+1)
print(palindrome(0))
```

```python
def fib_num(n):# multiple rec calls 
    if n==0 or n==1:
        return n
    return fib_num(n-1)+fib_num(n-2)
fib_num(7)
```
