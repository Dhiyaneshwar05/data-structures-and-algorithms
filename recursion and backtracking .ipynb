```python
def perm(nums):
    l=[]
    for i in range(0,len(nums)):
        for j in range(0,len(nums)):
            if i!=j:
                l.append([nums[i],nums[j]])
    """for k in l:
        k.append(nums not in k)"""
        
    return l

print(perm([1,2,3,4]))
```
